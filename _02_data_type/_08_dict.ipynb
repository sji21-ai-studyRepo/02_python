# dict(Dictionary, 사전 자료형)
- `dict`는 하나의 요소를 key : Value 형식으로 갖는 자료형이다.
- 사전 처럼 단어를 알면 뜻을 알 수 있듯이, Key를 통해 Value를 확인하는 구조
- list 처럼 index로 접근하는 것이 아닌 `key`로 접근 가능

## 1. dict 생성과 key 규칙

`dict`의 핵심은 `key`이다.

- 하나의 `dict` 안에서 `key`는 중복될 수 없다.
- 같은 `key`를 다시 넣으면 마지막 값으로 덮어써진다.
- `key`는 값을 빠르게 찾기 위한 기준이므로 중간에 값이 바뀌면 안 된다.
- 그래서 `int`, `float`, `str`, `tuple` 같은 불변 자료형은 `key`로 사용할 수 있지만, `list`, `dict`, `set` 같은 가변 자료형은 `key`로 사용할 수 없다.
- Python 3.7 이상에서 `dict`는 삽입 순서를 유지한다.

```python
from babel.plural import value_node
from prompt_toolkit import keys

print("--- dict ---")
dct = {
    'a' : 10,
    'b' : 20,
    'c' : 100
}

print(dct, type(dct))

# key는 불변 타입만 가능
dct2 = {
    10 : 'aaa',
    (10, 20) : 'bbb'
}
print(dct2, type(dct2))

# key 가변 타입 불가 확인
# dct3 = {
#     [1,2,3] : 'ccc'
# }
```

```python
print("--- dict 조회 ---")

print("dct: ", dct)

# key를 이용한 조회
print("dct['a']: ", dct['a'])
print("dct,get('a'): ", dct.get('a'))

# 없는 key 조회
# print("dct['d']: ", dct['d']) # KeyError 발생
print("dct,get('d'): ", dct.get('d')) # 없으면 None

# 없으면 기본값 적용
print("dct,get('d', '값 없음'): ", dct.get('d', '값 없음'))
print("dct,get('c', '값 없음'): ", dct.get('c', '값 없음'))
```

## dict 요소 추가, 수정, 삭제
- dict는 mutable 자료형
- dict[없는key] = value로 value 추가
- dict[중복key] = value로 value 수정 가능
- dict.update()를 이용해 value 수정 가능 (여러 값 동시 수정도 가능)
- pop() 또는 del을 이용해 요소 제거


```python
dct['e'] = 888 # 추가
print(dct)

dct['e'] = 999 # 수정
print(dct)

# update()
dct.update(e = 1234)
print(dct)

# 여러 value update
dct.update( {'a':111, 'b':222} )
print(dct)

# 요소 제거
# 1) del 연산
del dct['e']

# 2) pop() 메소드(꺼내기)
temp = dct.pop('c')
print(temp)
print(dct)
```

```python
# dict 생성 함수와 주요 메서드
# - dict 생성 방법
# 1) {k:v, k:v, ... }
# 2) dict(key=value, key=value, ...)
dct3 = dict(name="홍길동", age=22)
print(dct3)

# 주요 메서드
keys = dct3.keys()
print(keys, type(keys)) # dict_keys 타입

values = dct3.values() # 값만 추출
print(values, type(values)) # dict_values 타입

items = dct3.items() # item(요소 하나 == k:v) 추출
print(items, type(items)) # dict_items
```
