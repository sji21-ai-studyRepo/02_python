# Namespace

namespace는 변수, 함수, 클래스 등의 이름을 구분해서 관리하는 공간이다.
같은 이름이 있더라도 모듈, 클래스, 인스턴스라는 서로 다른 namespace에 있으면 충돌을 줄일 수 있다.


## 1. 모듈 namespace

`my_math.x`처럼 모듈 이름을 붙여 접근하면 현재 파일의 `x`와 사용자 모듈의 `x`를 구분할 수 있다.


```python
x = 10

from skn import my_math
print("x: ", x)
print("my_math.x: ", my_math.x)
```

## 2. 클래스 namespace와 인스턴스 namespace

클래스 속성은 클래스가 가진 이름 공간에 저장되고, 인스턴스 속성은 각 객체가 가진 이름 공간에 저장된다.


```python
import skn.my_math as my_math

class MyClass:
    x = 33 # 클래스 변수

    def __init__(self, x):
        self.x = x # self.x == 인스턴스 변수

print("MyClass.x (클래스 변수): ", MyClass.x)

my_clz = MyClass(55)
print("MyClass.x (인스턴스 변수): ", my_clz.x)
```

## 3. 정리

같은 이름 `x`라도 `x`, `my_math.x`, `MyClass.x`, `my_clz.x`는 서로 다른 namespace에 속할 수 있다.
namespace를 이해하면 import 후 이름 충돌과 객체 속성 접근 방식을 더 쉽게 설명할 수 있다.
